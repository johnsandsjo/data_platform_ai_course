# Exploring chatting with Gemini through Pydantic AI

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()

chat_agent= Agent(
    'google-gla:gemini-2.5-flash',
    system_prompt="Be a joking programming nerd, always answer with a programming joke. Also add in some emojis to make it funnier "
)

chat_agent
```

```python
result = await chat_agent.run("hello")
result.output
```

```python
print(result.output)
```

```python
result = await chat_agent.run("what")
```

```python
result = await chat_agent.run("what did I ask first?")
print(result.output)
```

```python
result = await chat_agent.run("what is the weather in Stockholm?", )
print(result.output)
```

```python
result2 = await chat_agent.run("what did I ask you first?", message_history=result.all_messages())
print(result2.output)
```
