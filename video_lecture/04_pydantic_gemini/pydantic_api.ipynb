## using pydantic
to dezerialize the json object into pydantic model

```python
#have a look at the data
data.get("results")[:5]

#list of dictionaries
```

```python
from pydantic import BaseModel

class PokemonListResponse(BaseModel):
    count: int
    results: list[dict]

pokemons = PokemonListResponse.model_validate(data)
pokemons
```

```python
#this is a class
type(pokemons)
```

```python
pokemons.count
```

```python
pokemons.results
```

```python
from pydantic import ValidationError
class PokemonListResponse(BaseModel):
    count: str
    results: list[dict]

try:
    pokemons = PokemonListResponse.model_validate(data)
except ValidationError as err:
    print(err)
```

```python
pikachu_data = requests.get("https://pokeapi.co/api/v2/pokemon-species/pikachu").json()
pikachu_data.keys()
```

```python
for key, value in pikachu_data.items():
    print(f"{key:<30} {type(value)}")
```

```python
class Pokemon(BaseModel):
    id: int
    name: str
    base_happiness: int
    capture_rate: int
    evolution_chain: dict
    gender_rate: int
    generation: dict

pikachu = Pokemon.model_validate(pikachu_data)
pikachu
```

```python
pikachu.evolution_chain
```

```python
pikachu.generation
```

```python
#here we serialize
serialized_pikachu = pikachu.model_dump_json()
serialized_pikachu
```

```python
type(serialized_pikachu)
```
