# 0. Summarize job ads
In the data folder, there are a few job ads taken from arbetsförmedlingen.se.

### a) Read all the jobs ads into python

```python
from pathlib import Path

current_dir = Path.cwd()
folderpath = current_dir / 'assets'
print(folderpath)
```

```python
ads_list = {}
current_dir = Path.cwd() 
target_folder = Path.cwd() / "assets"

for file_path in target_folder.glob('*.txt'): # Use glob('*.txt') to select only files ending in .txt
    with open(f"{file_path}") as f:
        ads_list[file_path.name] = f.read()
print(ads_list.keys())
#print(ads_list)
```

### b) Create a function that uses gemini to summarize a job ad. This function should take in an ad as its parameter and return a summary.

```python
from google import genai
client = genai.Client()
```

```python
from google.genai import types
def summarize_ads(prompt):
    system_instruction = "You will recieve a job ad, you should always reply with a very brief summary of it"
    response = client.models.generate_content(
    model = "gemini-2.0-flash",
    contents = prompt,
    config=types.GenerateContentConfig(system_instruction= system_instruction)

    )
    return response.text

prompt= ads_list["ads1.txt"]
summarize_ads(prompt)
```

### c) Now create and export markdown files for each job ad and its corresponding summary.

```python
def markdown_ads(prompt, file_name):
    system_instruction = """You will recieve a job ad in plain text. Always reply with a markdown file format. 
    The markdown file should include the summarized job ad on top with max 100 words, followed by the full add below"""
    response = client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = prompt,
    config=types.GenerateContentConfig(system_instruction= system_instruction)

    )
    with open(f"exports/{file_name}.md", "w") as file:
        return file.write(response.text)

prompt= ads_list["ads3.txt"]
markdown_ads(prompt, file_name="summarize4")
```
