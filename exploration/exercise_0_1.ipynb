```python
from pydantic import ValidationError
#basic basic example of how validation error works and see how pydantic may be casting the input

class Model(BaseModel):
    items: list[int]  

try:
    print(Model(items=(32873.34234, "2", "3")))
except ValidationError as e:
    print(e)
```

```python
#Gemini made up exercise
from pydantic import BaseModel, Field
class ProductOrder(BaseModel): 
    product_name: str
    price: float
    quantity: int = Field(gt=0)
    is_in_stock: bool = True

#Instantiate the ProductOrder model using valid data and print the resulting object.
productorder = ProductOrder(product_name="Boll", price=20.0, quantity=2, is_in_stock=True)
print(productorder)
```

```python
try:
    ProductOrder(product_name="Boll", price="ten sek", quantity=2, is_in_stock=True)
except ValidationError as e:
    print(e)
```

```python
try:
    ProductOrder(product_name="Boll", price=10.0, quantity=-1, is_in_stock=True)
except ValidationError as e:
    print(e)
```
